```python
import numpy as np 
import matplotlib.pyplot as plt 

# Frequencies
f = 10   # Original signal frequency (Hz)
f1 = 15  # Low sampling rate (Hz)
f2 = 150 # High sampling rate (Hz)

# Continuous time vector
t = np.linspace(0, 1, 1000)
x = np.sin(2 * np.pi * f * t)  # Original signal

# Discrete sample indices
n1 = np.arange(0, 1, 1/f1)
n2 = np.arange(0, 1, 1/f2)

# Sampled signals
x1 = np.sin(2 * np.pi * f * n1)  # Sampled at 15 Hz
x2 = np.sin(2 * np.pi * f * n2)  # Sampled at 150 Hz

# Aliased signal for comparison
alias_f = abs(f1 - f)
alias_signal = np.sin(2 * np.pi * alias_f * n1)

# Plotting
plt.figure(figsize=(14, 8))

# Original signal
plt.subplot(3, 1, 1)
plt.plot(t, x, label='Original Signal (10 Hz)', color='blue')
plt.title("Original Continuous-Time Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.axhline(0, color="red", ls='--')
plt.axvline(0, color="red", ls='--')
plt.legend()
plt.grid(True)

# Sampled at 15 Hz (Aliased)
plt.subplot(3, 1, 2)
plt.plot(n1, x1, label='Sampled at 15 Hz', color='blue')
plt.plot(n1, alias_signal, 'r--', label='Aliased Signal (|f - fₛ| = 5 Hz)')
plt.plot(t, x, 'k--', alpha=0.3, label='Original (reference)')
plt.stem(n1, x1, linefmt='C1-', markerfmt='C1o', basefmt=" ")
plt.title("Sampled at 15 Hz (Below Nyquist → Aliasing)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)

# Sampled at 150 Hz (Accurate)
plt.subplot(3, 1, 3)
plt.plot(n2, x2, label='Sampled at 150 Hz', color='green')
plt.stem(n2, x2, linefmt='C2-', markerfmt='C1o', basefmt=" ")
plt.plot(t, x, 'k--', alpha=0.3, label='Original (reference)')
plt.title("Sampled at 150 Hz (Above Nyquist → Accurate)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
