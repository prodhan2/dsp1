```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import firwin, convolve

# 1. Signal setup
fs = 1000  # Sampling frequency
t = np.arange(0, 1, 1/fs)
clean = np.sin(2 * np.pi * 10 * t)
noise = 0.3 * np.sin(2 * np.pi * 300 * t)  # reduced noise
noisy = clean + noise

# 2. FIR Low-pass filter with cutoff = 20 Hz
cutoff = 20  # in Hz
num_taps = 101  # filter length (more taps = smoother filter)
fir_coeff = firwin(num_taps, cutoff, fs=fs)

# 3. Apply FIR filter using convolution
filtered = convolve(noisy, fir_coeff, mode='same')

# 4. Plot all signals
plt.figure(figsize=(10, 8))

plt.subplot(3, 1, 1)
plt.plot(t, clean, color='green')
plt.title('Clean Signal')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(t, noisy, color='red')
plt.title('Noisy Signal (with 300Hz noise)')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(t, filtered, color='blue')
plt.title('Filtered Signal (Low-pass, 20 Hz cutoff)')
plt.xlabel('Time (s)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
