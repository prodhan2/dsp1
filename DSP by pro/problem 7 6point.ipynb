```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
n = np.arange(-10,21,0.3)
x = np.sin(0.5*np.pi*n)
N = len(x)
avg = np.zeros(N)
diff = np.zeros(N)

for i in range(5,N):
    avg[i] = (x[i]+x[i-1]+x[i-2]+x[i-3]+x[i-4]+x[i-5])/6
    diff[i] = (x[i]-x[i-1]+x[i-2]-x[i-3]+x[i-4]-x[i-5])/6
plt.figure(figsize=(10,5))

plt.subplot(3,1,1)

plt.plot(n,x , label= "original signal")
plt.plot(n,avg, label= "average signal")
plt.title("Original Signal")
plt.grid()
plt.legend()

plt.subplot(3,1,2)
plt.plot(n,x , label= "original signal")
plt.plot(n,diff, label= "average signal")
plt.title("Original Signal")
plt.grid()
plt.legend()

plt.subplot(3,1,3)
plt.plot(n,x , label= "original signal")
plt.plot(n,avg, label= "average signal")
plt.plot(n,diff, label= "average signal")

plt.title("Original Signal")
plt.grid()
plt.legend()


plt.tight_layout()

```
