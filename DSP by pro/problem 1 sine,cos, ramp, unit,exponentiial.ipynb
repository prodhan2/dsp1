```python
import numpy as np
import matplotlib.pyplot as  plt
np.random.seed(34)
```

Unit step signal : u(n) = 1 if n>=1
                         = 0  otherwise

```python
n = np.arange(-10,11,1)
un = np.where(n>=0 , 1, 0)
print(n)
print(un)
plt.stem(n,un,label="Unit stemp")
plt.legend()
plt.grid(True)
plt.show()
```

Ramp signal by sujan prodhan   u(n) = n if n>=0
       = 0 otherwise

```python
n = np.arange(-10,11,1)
rm = np.where(n>=0 , n , 0)
print(n)
print(rm)
plt.stem(n,rm,label= "Ramp signal ")
plt.legend()
plt.grid()
plt.show()
```

Exponential signal    y (t) = A e^bt 
A = 1 , b =0.5 , t = np.linespace(0,10,50)

```python
A = 1
b = 0.5
t = np.linspace(0,10,50)
# t = np.arange(-10,11,1)
y = A*np.exp(b*t)
plt.stem(t, y , label = "Exponential signal ")
plt.legend()
plt.xlabel("time")
plt.ylabel("Exponential growth")
plt.grid()
plt.show()
```

```python
x = np.linspace(0 , 2*np.pi, 1000)
A = np.cos(x)
B = np.sin(x)

plt.subplot(2,1,1)
plt.plot(x, A, label="coseWave")
plt.axvline(x =0 , color = "red")
plt.axhline(y=0, color = "green")
plt.grid()
plt.legend()


plt.subplot(2,1,2)
plt.plot(x, B, label="SineWave")
plt.grid()
plt.axvline(x =0, color = "red")
plt.axhline(y=0,color = "green")
plt.legend()
plt.show()


```
