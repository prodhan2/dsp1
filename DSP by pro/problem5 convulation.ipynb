```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
def unit(n):
 return  np.where(n>=0, 1, 0)

n = np.arange(0,11)
x = unit(n)
h = unit(n) - unit(n-5)
y = np.convolve(x,h)

ylen = np.arange(0,len(y))

plt.figure(figsize=(10, 5))
plt.subplot(3, 1, 1)
plt.stem(n, x )
plt.title('x[n]')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.grid()
plt.subplot(3, 1, 2)
plt.stem(n, h)
plt.title('h[n]')
plt.xlabel('n')
plt.ylabel('h[n]')
plt.grid()

plt.subplot(3, 1, 3)
plt.stem(ylen, y)
plt.title('y[n]')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.tight_layout()
plt.grid()
```
