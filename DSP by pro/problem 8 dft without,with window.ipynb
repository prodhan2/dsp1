```python
import numpy as np
import matplotlib.pyplot as plt

fs = 8000
N = 800
n = np.arange(N)
x = np.sin(2*np.pi*1000*n/fs) + 0.5*np.sin(2*np.pi*2000*n/fs)
w = np.hamming(N)
x_w = x *w

x = np.fft.fft(x)
xw = np.fft.fft(x_w)

freq = np.fft.fftfreq(N,1/fs)

plt.figure(figsize= (12,6))
plt.subplot(2,1,1)
plt.plot(freq[:N//2],np.abs(x[:N//2]))
plt.title("DFT WITH Window")
plt.grid()

plt.subplot(2,1,2)
plt.plot(freq[:N//2], np.abs(xw[:N//2]))
plt.title("DFT WITHOUT Window")
plt.grid()
plt.tight_layout()
plt.show()




```
