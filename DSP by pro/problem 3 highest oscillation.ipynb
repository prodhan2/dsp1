```python
import numpy as np
import matplotlib.pyplot as plt
omega1 = 0.5*np.pi
omega2 = np.pi
n = np.arange(0,11,1)
x1 = np.cos(omega1*n)
x2 = np.cos(omega2*n)

plt.figure(figsize= (10,5))
plt.subplot(2,1,1)
plt.title("omega  is 0.5 pi")
plt.stem(n,x1 , label = "lowest ocillation")
plt.legend()
plt.grid(True)

plt.figure(figsize= (10,5))
plt.subplot(2,1,1)
plt.title("omega is pi")
plt.stem(n,x2 , label = "Higest ocillation ")
plt.legend()
plt.grid(True)

plt.show()
```
