```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,3,-2,4])
y = np.array([1,3,-1,3])
z = np.array([2,-1,4,2])
rxy = np.correlate(x, y,mode ='full')
ryz = np.correlate(y, x, mode= 'full')
lags = np.arange(-len(x)+1 , len(x))

plt.figure(figsize=(8, 6))
plt.subplot(2, 1, 1)
plt.title('Cross-correlation of x and y')
plt.subplot(2, 1, 1)
plt.axvline(0, color='black', lw=1)
plt.axhline(0, color='black', lw=1)

plt.stem(lags, rxy )
plt.xlabel('Lags')
plt.ylabel('Cross-correlation')
plt.grid(True)



plt.subplot(2, 1, 2)
plt.title('Cross-correlation of y and z')
plt.stem(lags, ryz )
plt.xlabel('Lags')
plt.ylabel('Cross-correlation')
plt.axvline(0, color='black', lw=1)
plt.axhline(0, color='black', lw=1)
plt.grid(True)
plt.tight_layout()
plt.show()



```
