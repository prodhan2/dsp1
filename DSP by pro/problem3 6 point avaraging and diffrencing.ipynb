```python
import numpy as np
import matplotlib.pyplot as plt
n = np.arange(-10, 21, .3)
x = np.sin(.5*np.pi *n)
N = len(x)
y_avg = np.zeros(N)
y_diff = np.zeros(N)

for n in range(5, N):
    y_avg[n] = (x[n]+x[n-1]+x[n-2]+x[n-3]+x[n-4]+x[n-5])/6
    
for n  in range(5, N):
     y_diff[n] = (x[n]-x[n-1]-x[n-2]-x[n-3]-x[n-4]-x[n-5])/6
    
    
plt.figure(figsize=(10, 8))
plt.subplot(3,1,1)
plt.plot(x , label ="original signal x" ,  marker='o')
plt.plot(y_avg , label ="average signal y_avg" , linestyle='--', marker='o' ,linewidth=2)
plt.title("Average Signal")
plt.xlabel("n")
plt.ylabel("y_avg")
plt.grid(True)
plt.legend()
    
plt.subplot(3,1,2)
plt.plot(x , label ="original signal x" , marker='o')
plt.plot(y_diff , label ="diffrecenc signal  by y[n]" , linestyle='--', marker='o',linewidth=2)
plt.title("Average Signal")
plt.xlabel("n")
plt.ylabel("y_avg")
plt.grid(True)
plt.legend()
    
plt.subplot(3,1,3)
plt.plot(x , label ="original signal x" , marker='o')
plt.plot(y_avg , label ="average signal y_avg" , linestyle='--', marker='o' ,linewidth=2)

plt.plot(y_diff , label ="difference signal y_diff" , linestyle='--', marker='o',linewidth=2)
plt.title("Difference Signal")
plt.xlabel("n")
plt.ylabel("y_diff")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
    
    
```
